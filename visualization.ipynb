```python
import pandas as pd

# Load CSV file
df = pd.read_csv('experiments/experiment_v7/ResNet40_v3_training_history.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

# Plot 1: Train Loss, Test Loss on left axis; Learning Rate on right axis
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(df['epoch'], df['train_loss'], label='Train Loss', marker='o', color='tab:blue')
ax1.plot(df['epoch'], df['test_loss'], label='Test Loss', marker='o', color='tab:orange')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.legend(loc='upper left')
ax1.grid(True)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(df['epoch'], df['learning_rate'], label='Learning Rate', marker='o', color='tab:green')
ax2.set_ylabel('Learning Rate', color='tab:green')
ax2.tick_params(axis='y', labelcolor='tab:green')
ax2.legend(loc='upper right')

plt.title('Training and Testing Loss with Learning Rate')
plt.show()

# Plot 2: Train Accuracy, Test Accuracy on left axis; Learning Rate on right axis
fig, ax1 = plt.subplots(figsize=(10, 6))
line1, = ax1.plot(df['epoch'], df['train_accuracy'], 
                  label='Train Accuracy', 
                  color='tab:blue', 
                  linewidth=2)  # no marker
line2, = ax1.plot(df['epoch'], df['test_accuracy'], label='Test Accuracy', marker='o', color='tab:orange')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.grid(True)

ax2 = ax1.twinx()  # create a second axes sharing the x-axis
line3, = ax2.plot(df['epoch'], df['learning_rate'], label='Learning Rate', marker='o', color='tab:green')
ax2.set_ylabel('Learning Rate')
ax2.tick_params(axis='y')

# Combine legends from both axes
lines = [line1, line2, line3]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper right')

plt.title('Training and Testing Accuracy with Learning Rate')
plt.show()
```
